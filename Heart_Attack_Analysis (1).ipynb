```python
# Set up the matplotlib figure
plt.figure(figsize=(15, 12))
```

```python
# Age distribution
plt.subplot(3, 1, 1)
sns.histplot(heart_data, x='age', hue='output', multiple='stack', bins=20, kde=True)
plt.title('Age Distribution by Heart Attack Outcome')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
# Cholesterol distribution
plt.subplot(3, 1, 2)
sns.histplot(heart_data, x='chol', hue='output', multiple='stack', bins=20, kde=True)
plt.title('Cholesterol Distribution by Heart Attack Outcome')
plt.xlabel('Cholesterol')
plt.ylabel('Count')
```

```python
# Maximum heart rate distribution
plt.subplot(3, 1, 3)
sns.histplot(heart_data, x='thalachh', hue='output', multiple='stack', bins=20, kde=True)
plt.title('Maximum Heart Rate Distribution by Heart Attack Outcome')
plt.xlabel('Maximum Heart Rate')
plt.ylabel('Count')
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
# Correlation matrix
corr_matrix = heart_data.corr()
```

```python
# Set up the matplotlib figure for the correlation matrix
plt.figure(figsize=(12, 10))
```

```python
#Heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot to visualize relationships between features
# Due to the large size, we'll only select a few key features for visualization
sns.pairplot(heart_data[['age', 'trtbps', 'chol', 'thalachh', 'output']], hue='output', diag_kind='kde')
plt.show()
```

```python
# Integration with O2 Saturation Data
# Assuming the O2 Saturation data has the same length as the heart data and is in order
heart_data['o2_saturation'] = o2_saturation_data.iloc[:, 0]
```

```python
# Visualize O2 Saturation levels by heart attack outcome
plt.figure(figsize=(10, 6))
sns.histplot(heart_data, x='o2_saturation', hue='output', multiple='stack', bins=20, kde=True)
plt.title('O2 Saturation Levels by Heart Attack Outcome')
plt.xlabel('O2 Saturation')
plt.ylabel('Count')
plt.show()
```

```python
# Save the updated heart_data with O2 saturation for further analysis or model building
heart_data.to_csv('path_to_save_heart_data_with_o2.csv', index=False)
```
